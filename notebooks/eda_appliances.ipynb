# EDA: Energy Appliances Dataset

Exploratory Data Analysis for the Energy Appliances dataset with target: Appliances


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
df = pd.read_csv('../data/energydata_complete.csv')

# Basic info
print(df.info())
print(df.describe())
print(df.head())
```

```python
# Target distribution
plt.figure(figsize=(10, 6))
df['Appliances'].hist(bins=50)
plt.title('Distribution of Appliances Energy Consumption')
plt.xlabel('Appliances')
plt.ylabel('Frequency')
plt.show()
```

```python
# Correlation matrix (sample for large datasets)
if len(df.columns) > 20:
    # Sample columns for visualization
    sample_cols = df.select_dtypes(include=[np.number]).columns[:20]
    corr_data = df[sample_cols].corr()
else:
    corr_data = df.corr()

plt.figure(figsize=(12, 10))
sns.heatmap(corr_data, annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Correlation Matrix')
plt.tight_layout()
plt.show()
```
