# EDA: Energy Efficiency Dataset

Exploratory Data Analysis for the Energy Efficiency dataset with target: Heating Load


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
df = pd.read_csv('../data/energy_efficiency_data.csv')

# Basic info
print(df.info())
print(df.describe())
print(df.head())
```

```python
# Target distribution
plt.figure(figsize=(10, 6))
df['Heating Load'].hist(bins=50)
plt.title('Distribution of Heating Load')
plt.xlabel('Heating Load')
plt.ylabel('Frequency')
plt.show()
```

```python
# Correlation matrix
plt.figure(figsize=(12, 10))
sns.heatmap(df.corr(), annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Correlation Matrix')
plt.tight_layout()
plt.show()
```
